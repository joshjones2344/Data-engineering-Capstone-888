```python
indicators_selected = ['1.4.1 Net JHS Enrolment Rate']
```

```python
one_indicator = pd.DataFrame ()

for region in regions_selected:
    if len (indicators_selected) > 0:
        temp_region = sdg_data [sdg_data['Geolocation'] == region][['Year', indicators_selected [0]]]
    else:
        temp_region = sdg_data [sdg_data['Geolocation'] == region][['Year']]
    
    temp_region = pd.concat ([sdg_data['Geolocation'], temp_region], axis=1)
    
    temp_region = temp_region.dropna (thresh = len (indicators_selected) + 1)
    temp_region ['Year'] = temp_region ['Year'].astype('int')
    
    one_indicator = pd.concat([one_indicator, temp_region])
    
one_indicator = one_indicator.reset_index (drop = True)
one_indicator
```

```python
#on one-indicator dataframe (one_indicator) i want to add column indicator name and input the indicator name "1.4.1 Net JHS Enrolment Rate" to all rows in column indicator to last column of dataframe
one_indicator ['indicator'] = "1.4.1 Net JHS Enrolment Rate"
one_indicator
```

```python
#change column '1.4.1 Net JHS Enrolment Rate' to 'value'
one_indicator = one_indicator.rename(columns={'1.4.1 Net JHS Enrolment Rate': 'value'})
one_indicator
```

```python
#save one_indicator dataframe to csv file
one_indicator.to_csv ('data/1.4.1 Net JHS Enrolment Rate.csv', index = False)
```
