```python
import csv

# Input and output file paths
input_file = 'data/new data/region_area_updated.csv'
output_file = 'data/new data/Region_Area_Dimension.csv'

# Open both files
with open(input_file, 'r') as csv_input, open(output_file, 'w', newline='') as csv_output:
    reader = csv.reader(csv_input)
    writer = csv.writer(csv_output)

    # Read and write the header row
    header = next(reader)
    writer.writerow(['Region_Key'] + header)

    # Initialize the surrogate key
    surrogate_key = 101

    # Read and write each data row, and increment the surrogate key
    for row in reader:
        # Format the surrogate key as a 3-digit string
        surrogate_key_str = str(surrogate_key).zfill(3)
        writer.writerow([surrogate_key_str] + row)
        surrogate_key += 1
```

```python
#next we create time dimension table  only for year 2000 to 2022 with 1 year interval and has a sorrugate key
import pandas as pd
import numpy as np
import csv

# Create a list of years from 2000 to 2022
years = list(range(2000, 2023))

# Create a list of surrogate keys
surrogate_keys = list(range(1, len(years) + 1))

# Create a list of dictionaries, each containing a year and its surrogate key
year_keys = [{'Year': year, 'Year_Key': surrogate_key}
             for year, surrogate_key in zip(years, surrogate_keys)]

# Create a DataFrame from the list of dictionaries
year_dimension = pd.DataFrame(year_keys)
```

```python
#lets check the year_dimension
year_dimension.head()
```

```python
#swap the column
year_dimension = year_dimension[['Year_Key', 'Year']]
year_dimension.head()
```

```python
#change Year_Key to Time_Key
year_dimension = year_dimension.rename(columns={'Year_Key': 'Time_Key'})
year_dimension.head()
```

```python
#need to change the sorrogate key to 3 digit start in 701
year_dimension['Time_Key'] = year_dimension['Time_Key'] + 700
year_dimension.head()
```

```python
#save to new file
year_dimension.to_csv('data/new data/Time_Dimension.csv', index=False)
```

```python
merged_df = merged_df.merge(year_dimension[['Time_Key', 'Year']], on='Year', how='left')
```

```python
merged_df.head()
```
